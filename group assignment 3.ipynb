#### (A). Conditions for Calibration, FPR Parity, and FNR Parity

Calibration is about predictive meaning of the score while FPR/FNR parity is about distribution of harms.

- $\mathbb{P}$ - Is the probability.
- $Y = 1$ - A positive true outcome (for example the person actually did not recidivate(do crime again) or the loan should truly be approved).
- $S = s$ - Represents the continuous model score assigned by the algorithm (for example a risk score between 0 and 1).
- $A$: Represents the protected attribute (for example race, sex, or age).
    - $A = 0$ - The reference group (often the historically advantaged group, such as "White" or "Male" in typical auditing datasets).
    - $A = 1$ - The non-reference group (the protected or minority group being audited for bias).

Calibration is also known as Predictive Parity, A key group fairness definition in machine learning. Here probability of a positive true outcome given a specific model score $S=s$ is the same across all protected groups.

$$\mathbb{P}(Y=1|S=s,A=0)=\mathbb{P}(Y=1|S=s,A=1)$$

False Positive Rate (FPR) Parity: The probability of the model predicting a positive outcome when the true outcome is negative is equal across groups.

$$\mathbb{P}(\hat{Y}=1|Y=0,A=0)=\mathbb{P}(\hat{Y}=1|Y=0,A=1)$$

False Negative Rate (FNR) Parity: The probability of the model predicting a negative outcome when the true outcome is positive is equal across groups.
$$\mathbb{P}(\hat{Y}=0|Y=1,A=0)=\mathbb{P}(\hat{Y}=0|Y=1,A=1)$$

#### (B). Proof of the Impossibility Theorem

According to (Chouldechova, 2017; Kleinberg et al., 2016), The key identity linking calibration + FPR/FNR is

$$\text{PPV}_{g} = \frac{(1 - \text{FNR}_{g})\pi_{g}}{(1 - \text{FNR}_{g})\pi_{g} + \text{FPR}_{g}(1 - \pi_{g})}$$

- $\text{PPV}_{g}$: The Positive Predictive Value for group $g$, defined as the probability that the true outcome is positive given a positive prediction: $P(Y=1|\hat{Y}=1, A=g)$.
- $\text{FNR}_{g}$: The False Negative Rate for group $g$. Note that $(1 - \text{FNR}_{g})$ is equivalent to the True Positive Rate ($\text{TPR}$).
- $\text{FPR}_{g}$: The False Positive Rate for group $g$.
- $\pi_{g}$: The base rate (prevalence) for group $g$, representing the probability of a positive true outcome: $P(Y=1 | A=g)$.

There is need to show if calibration holds, PPV must be equal across groups but if base rates differ, the only way to equalize PPV is to change FPR/FNR.

$$\mathbb{P}(Y=1|A=0) \neq \mathbb{P}(Y=1|A=1)$$

Lastly there is need to conclude that all three cannot hold unless the classifier is trivial. No non-trivial classifier can simultaneously satisfy all three conditions. At most, only two of the three can hold simultaneously. This is not a technical bug but a mathematical reality of how these different fairness definitions relate to base rates. A trivial classifier = always predicts 0 or always predicts 1.

#### (D). Proposed Governance Rule

Responsible machine learning requires navigating three layers of alignment: mathematical (loss functions and metrics), behavioral (observed outputs), and institutional (deployment context and compliance). In practice, these layers often conflict due to the Impossibility Theorem.

The Core Conflict exist because:
- Calibration (Mathematical Alignment)- This ensures a model score carries the same predictive meaning across all groups. For example, a "High Risk" flag should correspond to the same probability of recidivism whether a defendant is Black or White.
- Error-Rate Parity (Institutional Alignment)- This governs the fair distribution of "accidental" harms, such as False Positives (wrongly flagging someone as high risk) or False Negatives (wrongly flagging someone as low risk).
- The Mathematical Reality - When base rates differ between groups, no non-trivial classifier can satisfy both calibration and equal error rates simultaneously.

Organizations must treat this conflict not as a technical bug, but as a value choice that must be explicitly documented and defended by:
1. Prioritization Strategy - An organization may prioritize calibration when the primary goal is accurate individual risk estimation. However, this choice is only defensible if business necessity can be clearly demonstrated.
2. The Burden Shifting Framework - Under legal and regulatory scrutiny (such as N.J. DCR Guidance), the operator bears the responsibility for auditing outputs. Even if business necessity is shown (Step 2), the organization must still prove that no less discriminatory alternative exists that could achieve similar results with lower disparity (Step 3).
3. Ongoing Stewardship - Because "fairness through blindness" fails—as removing protected attributes rarely removes their discriminatory effects—organizations must maintain rigorous subgroup testing and monitoring throughout the system's deployment.
