#### (a) Efficiency Axiom and Model "Race-Blindness"

##### State the Efficiency Axiom:
- The efficiency axiom requires that the sum of all feature contributions ($\phi_j$) must exactly equal the difference between the actual prediction for an instance $x$ and the average prediction across the entire dataset:

$$\sum_{j=1}^{p}\phi_{j}=\hat{f}(x)-\mathbb{E}[\hat{f}(x)]$$

The Efficiency Axiom states that the sum of the Shapley values (contributions) for all features must exactly equal the difference between the actual prediction for a specific instance and the average prediction for the entire dataset. The axiom ensures that every bit of the "payout" (the deviation from the average) is fully accounted for by the features; no credit is created or destroyed.

- $\phi_{j}$: The contribution (Shapley value) of feature $j$.
- $\hat{f}(x)$: The model's prediction for the specific individual $x$.
- $\mathbb{E}[\hat{f}(X)]$: The average (mean) prediction across all instances.

Why $\phi_{\text{race}} = 0$ implies "no contribution beyond the average":

The Shapley value $\phi_{race}$ measures how much the specific value of the "race" feature pushed the prediction away from the average. The Efficiency Axiom, the total prediction gap is shared among features. If $\phi_{race} = 0$, it simply means that for that specific instance, knowing the person's race provided zero additional information to the model to change the prediction from the baseline average. Hence the Shapley values distribute the "payout" (the prediction gap) among features. If $\phi_{\text{race}} = 0$, it means the specific value of the race feature for that instance did not push the prediction away from the mean.

On the contrary, A model is only "race-blind" if it does not use the race feature in its calculations at all. However, a model could be trained with race as an input and still result in $\phi_{race} = 0$ for a specific person if:

- The specific race of that individual happens to have an effect that perfectly aligns with the global average prediction.
- The model uses other features (proxies) that are correlated with race to make its decision, even if the race feature itself is assigned a zero contribution for that specific case.

#### (b) Two-Feature Example and COMPAS Interpretation

##### Constructing the Example:

Consider a model where:
- $x_1 = \text{race}$
- $x_2 = \text{prior arrests}$
- The model only uses $x_2$ to make predictions: $\hat{f}(x) = \beta_2 x_2$.

Model Setup: Imagine a linear model 

$$f(x) = 0 \cdot (race) + 1 \cdot (priors)$$

The Dummy Axiom of Shapley values dictates that a feature that never changes the prediction receives zero attribution. At the Individual Level, since $\beta_1 = 0$, the Shapley value $\phi_{\text{race}}$ will be exactly $0$ for every single defendant. At the Group Level, if the data shows that the Black group has a higher average number of prior arrests due to systemic factors, the model will flag Black defendants as high-risk more frequently than White defendants for the same underlying behavior.

Since the weight for race is zero, the Dummy Axiom ensures that $\phi_{race} = 0$ for every individual. However, if the data used to train the model has a higher base rate of prior arrests for one racial group due to systemic factors, the model will produce a higher False Positive Rate (FPR) for that group.

Interpretation in the COMPAS Context:
This result mirrors the COMPAS case where, even if a model is "blind" to race or assigns it no direct weight, it can still violate error-rate parity. This happens because the model uses features (like prior arrests) that act as proxies for race, leading to Black defendants having a higher average FPR than White defendants.

- Proxy Discrimination - In as much the model is interpretable by design because the weights are readable, yet it remains biased because the features ($X$) encode proxy discrimination.
- The Impossibility Theorem - The COMPAS case demonstrated that even when risk scores are calibrated across groups, error-rate parity ($FPR$ and $FNR$) often fails.
- Governance Failure - Relying solely on $\phi_{\text{race}} = 0$ as a fairness check is insufficient. It ignores the Clever Hans problem, where the model learns non-causal shortcuts (like systemic arrest patterns) that increase accuracy at the cost of being biased or fragile.

#### (c) Discussion: Sufficiency for Fairness

##### Is $\phi_{\text{race}} = 0$ a sufficient condition for fairness?

No. While $\phi_{\text{race}} = 0$ suggests that the model is not explicitly using race to differentiate an individual from the average, it is a local diagnostic tool, not a global guarantee of fairness. As seen in part (b), a model can have zero attribution for race but still produce racially biased outcomes through proxy discrimination or correlated features.

A Shapley value of zero for race only confirms that the model is "blind" to that specific column in your dataset. It fails to be a sufficient condition for fairness for several reasons:

- Proxy Discrimination - Models can learn to discriminate by using other features (like ZIP code or prior arrests) that are highly correlated with race, effectively reconstructing the sensitive attribute without directly "seeing" it.
- The Clever Hans Problem - High accuracy or "fair" attributions can hide the fact that a model is using non-causal shortcuts or spurious correlations that are actually proxies for protected groups.
- Adversarial Risk - Models can be intentionally engineered to produce "fair-looking" Shapley values while maintaining internal discriminatory logic.
- Transparency $\neq$ Ground Truth - Post-hoc methods like SHAP only approximate model behavior; they do not necessarily reveal the true causal internals of the system.

On the othet hand there is need for governance framework requirements. A robust governance framework should require both $\phi_{\text{race}} = 0$ and FPR (False Positive Rate) parity in high-stakes environments, such as criminal justice (COMPAS).

1. A $\phi_{\text{race}} = 0$ is a procedural fairness requirement since it will ensure "notice and opportunity to be heard" by demonstrating that an automated decision did not use a prohibited characteristic as a direct logical basis for its conclusion.

2. A FPR Parity is a distributive fairness or "outcome-based" requirement. In a governance framework. This will ensure that the "burden of error" is not disproportionately falling on one group. As seen in the COMPAS case, equal calibration is insufficient if one group is subjected to a much higher rate of false accusations (False Positives).

3. In conclusion combining the two requires both, a governance framework creates a multi-layered defense which acts as a Audit  Transparency tools (SHAP) which inspect the model's reasoning while the other arm will monitoring the workflows. Performance metrics (FPR parity) act as a safety check to ensure that even a "reasoning-blind" model isn't causing subgroup harm in production.
