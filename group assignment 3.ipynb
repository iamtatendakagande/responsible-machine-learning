#### (A). State the conditions for calibration, FPR parity, and FNR parity.

Calibration is about predictive meaning of the score while FPR/FNR parity is about distribution of harms.

- $\mathbb{P}$ - Is the probability.
- $Y = 1$ - A positive true outcome (for example the person actually did not recidivate(do crime again) or the loan should truly be approved).
- $S = s$ - Represents the continuous model score assigned by the algorithm (for example a risk score between 0 and 1).
- $A$: Represents the protected attribute (for example race, sex, or age).
    - $A = 0$ - The reference group (often the historically advantaged group, such as "White" or "Male" in typical auditing datasets).
    - $A = 1$ - The non-reference group (the protected or minority group being audited for bias).

Calibration is also known as Predictive Parity, A key group fairness definition in machine learning. Here probability of a positive true outcome given a specific model score $S=s$ is the same across all protected groups.

$$\mathbb{P}(Y=1|S=s,A=0)=\mathbb{P}(Y=1|S=s,A=1)$$

False Positive Rate (FPR) Parity: The probability of the model predicting a positive outcome when the true outcome is negative is equal across groups.

$$\mathbb{P}(\hat{Y}=1|Y=0,A=0)=\mathbb{P}(\hat{Y}=1|Y=0,A=1)$$

False Negative Rate (FNR) Parity: The probability of the model predicting a negative outcome when the true outcome is positive is equal across groups.
$$\mathbb{P}(\hat{Y}=0|Y=1,A=0)=\mathbb{P}(\hat{Y}=0|Y=1,A=1)$$

#### (B). Prove that when $\mathbb{P}(Y=1|A=0) \neq \mathbb{P}(Y=1|A=1)$, no non-trivial classifier satisfies all three simultaneously.

According to (Chouldechova, 2017; Kleinberg et al., 2016), The key identity linking calibration + FPR/FNR is

$$\text{PPV}_{g} = \frac{(1 - \text{FNR}_{g})\pi_{g}}{(1 - \text{FNR}_{g})\pi_{g} + \text{FPR}_{g}(1 - \pi_{g})}$$

- $\text{PPV}_{g}$: The Positive Predictive Value for group $g$, defined as the probability that the true outcome is positive given a positive prediction: $P(Y=1|\hat{Y}=1, A=g)$.
- $\text{FNR}_{g}$: The False Negative Rate for group $g$. Note that $(1 - \text{FNR}_{g})$ is equivalent to the True Positive Rate ($\text{TPR}$).
- $\text{FPR}_{g}$: The False Positive Rate for group $g$.
- $\pi_{g}$: The base rate (prevalence) for group $g$, representing the probability of a positive true outcome: $P(Y=1 | A=g)$.

There is need to show if calibration holds, PPV must be equal across groups but if base rates differ, the only way to equalize PPV is to change FPR/FNR.

$$\mathbb{P}(Y=1|A=0) \neq \mathbb{P}(Y=1|A=1)$$

1. The Proof Steps
To prove it, we need show the algebraic tension. If we assume the three fairness criteria hold we should:
    - Equalized Odds: $\text{FPR}_0 = \text{FPR}_1$ and $\text{FNR}_0 = \text{FNR}_1$.
    - Calibration: $\text{PPV}_0 = \text{PPV}_1$.

2. Step-by-Step Logic
    - Set $\text{PPV}_0 = \text{PPV}_1$.
    - Substitute the values of $\text{FPR}$ and $\text{FNR}$ since they are equal across groups.
    - Show that if $\text{PPV}_0 = \text{PPV}_1$:

    $$\frac{(1 - \text{FNR})\pi_0}{(1 - \text{FNR})\pi_0 + \text{FPR}(1 - \pi_0)} = \frac{(1 - \text{FNR})\pi_1}{(1 - \text{FNR})\pi_1 + \text{FPR}(1 - \pi_1)}$$

    -  The Contradiction: If we cross-multiply and simplify this, the equation will only holds if:

        - $\pi_0 = \pi_1$ (But the prompt says $\pi_0 \neq \pi_1$), OR
        - $\text{FPR} = 0$ and $\text{FNR} = 0$ (A perfect classifier, usually impossible in practice), OR
        - The classifier is trivial (e.g., if $\text{FPR} = 1 - \text{FNR}$, which happens when $\hat{Y}$ is independent of $Y$).


Lastly there is need to conclude that all three cannot hold unless the classifier is trivial. No non-trivial classifier can simultaneously satisfy all three conditions. In a trivial (random or constant) classifier, the True Positive Rate equals the False Positive Rate ($\text{TPR} = \text{FPR}$, or $1 - \text{FNR} = \text{FPR}$). At most, only two of the three can hold simultaneously. This is not a technical bug but a mathematical reality of how these different fairness definitions relate to base rates. A trivial classifier = always predicts 0 or always predicts 1.

#### (D). Propose a governance rule: under what circumstances should an organization prioritize calibration over FPR parity? Justify using the burden shifting framework.

Responsible machine learning requires navigating three layers of alignment which are mathematical (loss functions and metrics), behavioral (observed outputs), and institutional (deployment context and compliance). In practice, these layers often conflict due to the Impossibility Theorem.

An organization should prioritize Calibration (Sufficiency) over Error-Rate Parity (Separation) in scenarios where the model’s primary function is to provide individualized risk or benefit assessments that require consistent predictive meaning across groups. For example, in a medical diagnostic setting, a 70% risk must trigger the same clinical intervention regardless of a patient's race to ensure procedural fairness. In these high-stakes individual benefit contexts, prioritizing precision at the individual level is a business and ethical necessity to ensure equal treatment for equal risk.

Using the Burden-Shifting Framework, this prioritization is justified through a three-stage audit. First, the organization must acknowledge that if base rates ($\pi_g$) differ, prioritizing calibration will mathematically necessitate a Disparate Impact in False Positive Rates. Second, the operator must demonstrate Business Necessity; they must prove that failing to maintain calibration would compromise the model’s utility or lead to the misallocation of critical resources. Finally, the organization bears the burden of proving that they have searched for Less Discriminatory Alternative models or feature sets that could achieve similar calibration with a smaller gap in error rates.

Ultimately, navigating the Impossibility Theorem is not a technical bug but a value choice. While calibration is defensible for individualized care, Error Rate Parity should remain the priority in punitive or gatekeeping scenarios (e.g., criminal sentencing or basic employment) where the societal harm of a False Positive outweighs the need for precise individual probability. Responsible stewardship requires explicit documentation of these trade offs and continuous subgroup monitoring to ensure the chosen alignment remains ethically sound.
