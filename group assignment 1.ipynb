- $\mathcal{H}$ - It is the collection of every possible strategy or model the computer is allowed to choose from.
- $h$ - Hypothesis (or a candidate model). This is one specific choice from that menu. It’s a single candidate model or a specific "plan" the computer might pick.
- $L(h)$ - Proxy Loss function (or Objective Function) evaluated for a specific hypothesis $h$. For example, Since we can't easily measure "happiness," we give the computer a measurable goal instead (e.g., "Minimize the number of clicks it takes to buy something"). The computer tries to make this number as small as possible.
- $U(h)$ - The true human utility or intended objective. This represents what the human decision-maker actually cares about and wants to maximize. Hence this is the "Heart's Desire", It represents what the human actually cares about for example "I want my customers to be truly happy and loyal"). This is often very hard to turn into math.
- $h_L^\star$ - The hypothesis that is optimal under the proxy objective, i.e., the minimizer of $L(h)$. It represents (The Computer's Pick)which is the specific strategy that best solves the math shortcut. It’s the "winner" according to the proxy loss $L$.
- $h_U^\star$ - The hypothesis that is optimal under the intended human objective, i.e., the maximizer of $U(h)$. It represents (The Human's Dream) whuich is the specific strategy that would actually make humans the happiest. It’s the "winner" according to the true utility $U$.


## (a) Formal condition for alignment of the optimizer

$$\arg \min_{h \in \mathcal{H}} L(h) = \arg \max_{h \in \mathcal{H}} U(h)$$

$U(h)$ must be a strictly monotonically decreasing transformation of $L(h)$. Mathematically, this means there exists a function $f$ such that:$$U(h) = f(L(h)) \quad \text{where} \quad \frac{df}{dL} < 0$$

Its often encounter this when translating a "real world" goal into a "solver" goal:
- The Human Utility ($U$): Maximize Total Profit. 
- The Proxy Loss ($L$): Minimize Total Cost (where Cost = -Profit).
- The Result: Since minimizing costs is the same as maximizing the negative of those costs, the specific decision $h$ (like where to place a warehouse) will be identical for both.

If the condition fails?If $L(h)$ is not perfectly aligned with $U(h)$, we run into Objective Mismatch.

## (b) Construct an explicit counterexample: define a simple hypothesis space Hand explicit functions L(h) and U(h) for which the argmin and argmax differ.

To construct an explicit counterexample, we need to define a hypothesis space $\mathcal{H}$ and two functions where the "best" choice for the proxy loss ($L$) is a "bad" choice for the true human utility ($U$).

Let's define a simple scenario with only two possible models:
1. Hypothesis Space: $\mathcal{H} = \{h_1, h_2\}$
2. Proxy Loss $L(h)$: (What the computer minimizes)
$$L(h_1) = 10 $$
$$L(h_2) = 20$$
3. True Utility $U(h)$: (What the human actually wants)
$$U(h_1) = 50 $$
$$U(h_2) = 100 $$

The Resulting MismatchThe Computer's Choice ($h_L^\star$):

1. The Computer's Choice ($h_L^\star$):
The system looks for the minimum of $L(h)$. Since $10 < 20$, the optimal hypothesis is:
$$h_L^\star = \arg \min_{h \in \mathcal{H}} L(h) = h_1$$
2. The Human's Choice ($h_U^\star$):
The human looks for the maximum of $U(h)$. Since $100 > 50$, the optimal hypothesis is:
$$h_U^\star = \arg \max_{h \in \mathcal{H}} U(h) = h_2$$
3. Conclusion: In this case, $h_L^\star \neq h_U^\star$. The computer chose $h_1$ because it had a lower "loss," but $h_2$ was actually twice as useful to the human.

Conclusion

This happened because the formal condition from part (a) was violated. $U(h)$ is strictly increasing with respect to $L(h)$ in this example (as loss goes up, utility also goes up).

## (c) Interpret your example in a real-world ML setting (e.g., credit, hiring, risk scoring): identify the proxy objective and the intended human objective.

To interpret the counterexample in a real-world setting in Automated Hiring (Resume Screening). This is a classic area where proxy objectives often diverge from human utility. In this setting, a company uses an ML model to rank thousands of job applicants to decide whom to interview.

1. The Proxy Objective: $L(h)$ The "Loss" function the computer is told to minimize is often Predicted Tenure Error or Likelihood of a "Yes" from a Human Recruiter.
- The Data: The model is trained on historical data of who was hired in the past.
- The Objective ($L$): Minimize the difference between the model's prediction and the historical hiring decisions.
- The Result ($h_L^\star$): The model identifies $h_L^\star$, a hypothesis that favors candidates who "look" like previous hires (e.g., went to the same five universities or used specific keywords like "synergy").

2. The Intended Human Objective: $U(h)$ : The "True Utility" the CEO or Department Head actually wants to maximize is Long-term Innovation and Team Performance.

- The Goal ($U$): Find candidates who bring fresh perspectives, high problem-solving ability, and diverse skill sets that the company currently lacks.
- The Conflict: A candidate ($h_2$) might be a self-taught genius from a non-traditional background. They have a high $U(h)$ because they would revolutionize the team, but they have a high $L(h)$ because the model sees them as "risky" or "dissimilar" to the historical training data.

3. Conclusion:

- The goal on from from Proxy Objective $L(h)$ is to Minimize Prediction Error (Imitation) while the Intended Utility $U(h)$ is to maximise Maximize Organizational Value (Innovation).
- The Metric used for Proxy Objective $L(h)$ is Historical Correlation and the Intended Utility $U(h)$ is to Future Performance & Diversity.
